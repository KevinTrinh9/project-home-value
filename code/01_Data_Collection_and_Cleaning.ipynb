```python
# drop columns with more than 50% null values
threshold = 0.5 * len(train)
train_cleaned = train.dropna(thresh=threshold, axis=1)
```

```python
train_cleaned.info()
```

```python
# drop columns with more than 50% null values
threshold = 0.5 * len(test)
test_cleaned = test.dropna(thresh=threshold, axis=1)
```

```python
test_cleaned.info()
```

```python
# fill missing data in numerical columns with average of columns' values
# fill missing data in categorical columns with 'Unknown'
numerical_columns = train_cleaned.select_dtypes(include=['int', 'float']).columns
train_cleaned.loc[:, numerical_columns] = train_cleaned.loc[:, numerical_columns].fillna(train_cleaned[numerical_columns].mean())
categorical_columns = train_cleaned.select_dtypes(include=['object']).columns
train_cleaned.loc[:, categorical_columns] = train_cleaned.loc[:, categorical_columns].fillna('Unknown')
train_cleaned.info()
```

```python
# fill missing data in numerical columns with average of columns' values
# fill missing data in categorical columns with 'Unknown'
numerical_columns = test_cleaned.select_dtypes(include=['int', 'float']).columns
test_cleaned.loc[:, numerical_columns] = test_cleaned.loc[:, numerical_columns].fillna(test_cleaned[numerical_columns].mean())
categorical_columns = test_cleaned.select_dtypes(include=['object']).columns
test_cleaned.loc[:, categorical_columns] = test_cleaned.loc[:, categorical_columns].fillna('Unknown')
test_cleaned.info()
```
